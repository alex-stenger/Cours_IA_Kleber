Mais ce genre de dataset ne se prête pas du tout à la classification (quand on voit des données comme ça, on se dit directement qu'on va faire une régression)...

Regardons sur un jeu de données plus taillé pour la classification 

```python
x, y = make_blobs(n_samples=200, centers=3, n_features=2, random_state=9)
x+=15

fig, ax = plt.subplots()

scatter = ax.scatter(x[:,0],x[:,1], c=y, cmap="viridis")

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Classes")
ax.add_artist(legend1)
ax.set_xlabel("Largeur")
ax.set_ylabel("Longeur")
plt.title("Prédiction du fruit en fonction de sa longeur et sa largeur")
plt.show()
```
