## Génération aléatoire de données

```python
import sklearn
from sklearn import datasets
from sklearn.datasets import make_blobs
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = sklearn.datasets.make_regression(n_samples=100, n_features=1, noise=14, random_state=13)
x+=2.2
y+=70
x*=10
plt.scatter(x,y)
plt.show()

print(x[:,0].shape)

df = pandas.DataFrame(data={"c1": x[:,0], 
                            "c2": y})
df.to_csv("data/data2.csv", sep=',',index=False)
```

```python
x, y = make_blobs(n_samples=200, centers=3, n_features=2, random_state=9)
x+=15

# courge
# orange
# pastèque
# Attention on a deux variables cette fois-ci, la hauteur et la largeur du fruit/legume.


plt.scatter(x[:,0],x[:,1], c=y)
plt.xlabel("largeur")
plt.ylabel("hauteur")
plt.show()
```

```python
x, y = make_blobs(n_samples=200, centers=3, n_features=2, random_state=9)
x+=15

fig, ax = plt.subplots()

scatter = ax.scatter(x[:,0],x[:,1], c=y)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Classes")
ax.add_artist(legend1)
ax.set_xlabel("Largeur")
ax.set_ylabel("Longeur")
plt.title("Prédiction du fruit en fonction de sa longeur et sa largeur")
plt.show()
```
